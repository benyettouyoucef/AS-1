**Exercice : Add Two Tensors**

*1. Important note!*

**Lua compares tables, userdata, and functions by reference, that is, two such values are considered equal only if they are the very same object. For instance, after the code**

```lua
function addTensors(a,b)
    local res = a:add(b)
    return res
end
```

```lua
x = torch.Tensor(2, 2):fill(2)
y = torch.Tensor(4):fill(3)
```

```lua
function check_tensor_equality(a,b)
    if a:nDimension() ~= b:nDimension() then
        return "Tensor doesn't have same dimension ==> not equals" 
    end
    for i=1,a:nDimension() do
        for j=1,#a[i]:size() do
            if a[i][j] ~= b[i][j] then print(a[i][j]) print(b[i][j]) return "tensors not equal" end
        end
    end
    return "Tensors are equal (congratsss)"
end
```

```lua
submition = addTensors(x,y)
x = torch.Tensor(2, 2):fill(2)
y = torch.Tensor(4):fill(3)
test = torch.add(x,y)

--[[ Checking Tensors equality]]

check_tensor_equality(submition,test)
```

## Neural Networks

```lua
require 'nn';
```

### Examples of a NN

```lua
net = nn.Sequential()
net:add(nn.SpatialConvolution(1, 6, 5, 5)) -- 1 input image channel, 6 output channels, 5x5 convolution kernel
net:add(nn.ReLU())                       -- non-linearity 
net:add(nn.SpatialMaxPooling(2,2,2,2))     -- A max-pooling operation that looks at 2x2 windows and finds the max.
net:add(nn.SpatialConvolution(6, 16, 5, 5))
net:add(nn.ReLU())                       -- non-linearity 
net:add(nn.SpatialMaxPooling(2,2,2,2))
net:add(nn.View(16*5*5))                    -- reshapes from a 3D tensor of 16x5x5 into 1D tensor of 16*5*5
net:add(nn.Linear(16*5*5, 120))             -- fully connected layer (matrix multiplication between input and weights)
net:add(nn.ReLU())                       -- non-linearity 
net:add(nn.Linear(120, 84))
net:add(nn.ReLU())                       -- non-linearity 
net:add(nn.Linear(84, 15))                   -- 15 is the number of outputs of the network (in this case, 15 digits)
net:add(nn.LogSoftMax())                     -- converts the output to a log-probability. Useful for classification problems
```

```lua
print('Lenet5\n' .. net:__tostring());
```

### Applying my very first neural network(!!!)

```lua
input = torch.rand(1,32,32)
output = net:forward(input)
net:zeroGradParameters() -- zero the internal gradient buffers of the network (will come to this later)
gradInput = net:backward(input,torch.rand(15))
```

```lua
print(output)
```

## Loss function implementation

```lua
criterion = nn.ClassNLLCriterion() -- Negative Log-Likelihood Criterion
criterion:forward(output,3) -- The second argument means that it is the true class
```

```lua
gradients = criterion:backward(output,3)
```

```lua
gradInput = net:backward(input, gradients)
```

#### Working Implementation of the Neural Network

```lua
require "nn"
mlp = nn.Sequential();  -- make a multi-layer perceptron
inputs = 2; outputs = 1; HUs = 20; -- parameters
mlp:add(nn.Linear(inputs, HUs))
mlp:add(nn.Tanh())
mlp:add(nn.Linear(HUs, outputs))
```

```lua
for i = 1,2500 do
  -- random sample
  local input= torch.randn(2);     -- normally distributed example in 2d
  local output= torch.Tensor(1);
  if input[1]*input[2] > 0 then  -- calculate label for XOR function
    output[1] = -1
  else
    output[1] = 1
  end

  -- feed it to the neural network and the criterion
  criterion:forward(mlp:forward(input), output)

  -- train over this example in 3 steps
  -- (1) zero the accumulation of the gradients
  mlp:zeroGradParameters()
  -- (2) accumulate gradients
  mlp:backward(input, criterion:backward(mlp.output, output))
  -- (3) update parameters with a 0.01 learning rate
  mlp:updateParameters(0.01)
end
```

```lua
----------------------------------------------------
-- 4. Modules.
----------------------------------------------------



--[[ I'm commenting out this section so the rest of
--   this script remains runnable.
-- Suppose the file mod.lua looks like this: ]]--
local LinearModule = {}

local function sayMyName()
  print('Hrunkner')
end

function LinearModule.sayHello()
  print('Why hello there')
  sayMyName()
end

return LinearModule
```

```lua

-- Another file can use mod.lua's functionality:
local mod = require('mod')  -- Run the file mod.lua.

-- require is the standard way to include modules.
-- require acts like:     (if not cached; see below)
local mod = (function ()
  <contents of mod.lua>
end)()
-- It's like mod.lua is a function body, so that
-- locals inside mod.lua are invisible outside it.

-- This works because mod here = M in mod.lua:
mod.sayHello()  -- Says hello to Hrunkner.
```
